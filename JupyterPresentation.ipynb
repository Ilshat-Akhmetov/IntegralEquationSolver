### $$ y(x) = sin(\pi*x) + \frac{1}{2}\int_{0}^1y(t)dt $$ ###
### $$ true\_sol = sin(\pi x)+\frac{2}{\pi}$$  ###

```python
right_part_of_eq = lambda x_var, target_func: target_func(x_var) - torch.sin(
        pi * x_var
)
integral_part = lambda x_curr_var, integral_domain_var,  target_func: 0.5 * target_func(integral_domain_var)
true_solution = lambda x_var: torch.sin(pi * x_var) + 2 / pi

IES = IntegralEquationSolver(0, 1, right_part_of_eq, integral_part, true_solution)
IES.train()
```

```python
IES.make_report()
```

### $$y(x) = 2 - 3\int_{0}^{\frac{\pi}{2}}sin(x-2t)y(t)dt$$ ### 
### $$true\_sol = 2 - 3 sin(x)$$ ###

```python
right_part_of_eq = lambda x_var, target_func: target_func(x_var) - 2

integral_part = lambda x_curr_var, integral_domain_var,  target_func: -3 * target_func(integral_domain_var) * torch.sin(x_curr_var - 2 * integral_domain_var)
true_solution = lambda x_var: 2 - 3 * torch.sin(x_var)
IES = IntegralEquationSolver(0, pi/2, right_part_of_eq, integral_part, true_solution)
IES.train()
```

```python
IES.make_report()
```

### $$ y(x) =  \int_{0}^{1}(\sqrt{xt}y(t)dt + x$$ ###
### $$ true\_sol =  x + \frac{4}{5}\sqrt{x}$$ ###

```python
right_part_of_eq = lambda x_var, target_func: target_func(x_var) - x_var

integral_part = lambda x_curr_var, integral_domain_var,  target_func: target_func(integral_domain_var) * torch.sqrt(x_curr_var * integral_domain_var)
true_solution = lambda x_var: x_var + 4/5*torch.sqrt(x_var)
IES = IntegralEquationSolver(0, 1, right_part_of_eq, integral_part, true_solution)
IES.train()
```

```python
IES.make_report()
```
